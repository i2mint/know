# stream2py and creek review

```python
import iplantuml  # pip install iplantuml
# See https://plantuml.com/ for examples
```

```python
%%plantuml

@startuml

actor User

User -> Ctrl: get_iterator
    
Ctrl -> BufferCtrl: start
queue Buffer

group Sourcing Buffer [loop until exception]
    BufferCtrl -> Source: read
    Source -> BufferCtrl: item

    BufferCtrl -> Buffer: write item
    BufferCtrl -> Pacer: item
    Pacer -> Pacer: sleepf
end
    
group BufferReader iterator (independent loop)
    Ctrl -> BufferCtrl: read
    BufferCtrl -> Buffer: read
    Buffer -> BufferCtrl: item
    BufferCtrl -> Ctrl: item
end
    
@enduml
```

```python
%%plantuml

@startuml
actor User

User -> Ctrl: get_wf_annots_iter
database Annots
database Waveform
Ctrl -> Annots: get_annots_iter
Annots -> Ctrl: annots

loop for each annot of annots
    Ctrl -> Waveform: get_slice(annot)
    Waveform -> Ctrl: wf_segment
end

Ctrl -> User: (annot, wf_segment)
@enduml
```
